# Matplot Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Line Plot

```python
# 1)  Plot a simple line graph showing sales vs year.
       # Add labels, title, legend to plot.
```

```python
sales = [25, 30, 50, 90, 80, 65, 55, 45, 60, 75]
year = np.arange(2001,2011,1)
plt.axis([2000,2011,0,100])
plt.xlabel("Year", fontsize=20)
plt.ylabel("Sales", fontsize=20)
plt.title("Sales Forecasting")
plt.plot(year, sales, color='y', linestyle= '--', linewidth=3, marker='*',ms=10, markeredgecolor='r', markerfacecolor='g')
plt.legend(["Sales Forecasting"], loc=1)
plt.show()
```

```python
 # use of grid 
```

```python
from matplotlib import style
style.use('ggplot')

sales = [25, 30, 50, 90, 80, 65, 55, 45, 60, 75]
year = np.arange(2001,2011,1)
plt.axis([2000,2011,0,100])
plt.xlabel("Year", fontsize=20)
plt.ylabel("Sales", fontsize=20)
plt.title("Sales Forecasting")
plt.plot(year, sales, color='y', linestyle= '--', linewidth=3, marker='*',ms=10, markeredgecolor='r', markerfacecolor='g')
plt.legend(["Sales Forecasting"], loc=1)

plt.grid(color='b',linewidth=1)
plt.show()
```

```python
# Plot multiple lines on the same graph 
```

```python
style.use('default') # to remove grid

sales1 = [25, 30, 50, 90, 80, 65, 55, 45, 60, 75]
sales2 = [30, 28, 55, 80, 35, 60, 45, 40, 70, 90]
sales3 = [35, 30, 60, 75, 40, 70, 50, 48, 80, 95]
year = np.arange(2001,2011,1)

plt.axis([2000,2011,0,100])
plt.xlabel("Year", fontsize=20)
plt.ylabel("Sales", fontsize=20)
plt.title("Sales Forecasting")

plt.plot(year, sales1, color='y', marker='*', label='sales_1')
plt.plot(year, sales2, color='r', marker='^', label='sales_2')
plt.plot(year, sales3, color='g', marker='o', label='sales_3')

plt.legend(loc=4)
plt.show()
```

```python
# type of markers '''
''' 'o'   circle  
    '*'   star  
    '^'   triangle up  
    's'   square  
    'D'   diamond  
    '>'   triangle right  
    '<'   triangle left  
    'p'   pentagon  
    'h'   hexagon '''  
```

# Bar Chart

```python
# 2) Plot a bar chart of highest 5 temperatures recorded in a week
```

```python
sales = [25, 30, 50, 90, 80, 65, 55, 45, 60, 75]
year = np.arange(2001,2011,1)

plt.xlabel("Year", fontsize=20)
plt.ylabel("Sales", fontsize=20)
plt.title("Sales Forecasting")

plt.bar(year, sales, width=0.3, color='y', align='edge', edgecolor='r') # use width in vertical bar chart

plt.legend(['sales'],loc=1)
plt.show()
```

```python
sales = [25, 30, 50, 90, 80, 65, 55, 45, 60, 75]
year = np.arange(2001,2011,1)

plt.ylabel("Year", fontsize=20)
plt.xlabel("Sales", fontsize=20)
plt.title("Sales Forecasting")

plt.barh(year, sales, height=0.3, color='y', align='edge', edgecolor='r')  # use height in horizontal bar chart

plt.legend(['sales'],loc=1)
plt.show()
```

# Scatter Plot

```python
sales1 = [25, 30, 50, 90, 80, 65, 55, 45, 60, 75]
sales2 = [30, 28, 55, 80, 35, 60, 45, 40, 70, 90]
year1 = np.arange(2001,2011,1)


plt.xlabel("Year", fontsize=20)
plt.ylabel("Sales", fontsize=20)
plt.title("Sales Forecasting")

plt.scatter(year1, sales1, color='r',marker='o',  label='sales_1')
plt.scatter(year1, sales2, color='g',marker='*',  label='sales_2')

plt.legend(loc=2)
plt.show()
```

# Pie Chart

```python
student = [ 65, 50, 75, 80, 90 ]
label = ['maths', 'english', 'science', 'hindi', 'computers']
sizes=[0.09,0.1,0.1,0.2,0.1]
clr =['c','y','m','g','r']

plt.pie(student, labels=label, explode=sizes, colors=clr)
plt.show()
```

```python
student = [ 65, 50, 75, 80, 90 ]
label = ['maths', 'english', 'science', 'hindi', 'computers']

sizes=[0.09,0.1,0.1,0.2,0.1]
clr =['c','y','m','g','r']
font = {'fontsize':15}
props = {"linewidth":4,"width":1,"edgecolor":"b"}
plt.pie(student, labels=label, explode=sizes, colors=clr, autopct="%0.2f%%", radius=2, shadow=True, textprops=font, wedgeprops=props, pctdistance=0.8)
plt.legend()
plt.show()
```

# SubPlots

```python
sales = [25, 30, 50, 90, 80, 65, 55, 45, 60, 75]
sales2 = [30, 28, 55, 80, 35, 60, 45, 40, 70, 90]
year = np.arange(2001,2011,1)
student = [ 65, 50, 75, 80, 90 ]
label = ['maths', 'english', 'science', 'hindi', 'computers']
```

```python
plt.figure(figsize=(15,12))

plt.subplot(2,3,1)
plt.plot(year,sales)
plt.title("Line Plot")

plt.subplot(2,3,2)
plt.bar(year,sales)
plt.title("Bar-chart")

plt.subplot(2,3,3)
plt.scatter(year,sales,color='b',marker='o')
plt.scatter(year,sales2,color='r',marker='*')
plt.title("Scatter-chart")

plt.subplot(2,3,4)
sizes=[0.09,0.1,0.1,0.2,0.1]
clr =['c','y','m','g','r']
plt.pie(student, labels=label, explode=sizes, colors=clr)
plt.title("Pie-Chart")

plt.subplot(2,3,5)
plt.barh(year,sales)
plt.title("Horizontal-Bar")

plt.subplot(2,3,6)
bins=np.arange(0,100,10)
plt.hist(sales,bins=bins,rwidth=0.2)
plt.title("Histogram")

plt.tight_layout()
plt.show()
```
