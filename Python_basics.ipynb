# 2) Write a function to check if a number is prime.

```python
num = int(input("Enter the number : "))

if (num==1):
    print(f'No, {num} is not a Prime Number')
elif (num==2):
    print(f'Yes, {num} is a Prime Number')
else:
    n = num
    count = 0
    for i in range(2,n,1):
        if(num%i==0):
            break
        else:
            count = count+1
    if(count == n-2):
        print(f'Yes, {num} is a Prime Number')
    else:
        print(f'No, {num} is not a Prime Number')
```

# 4) Write the Factorial value of an integer. 

```python
def fact(n):
    if n==0 or n==1 :
        return 1
    else:
        return (n * fact(n-1))

num = int(input("Enter the value to find its factorial value : "))
result = fact(num)
print(f'The factorial value of {num} is :', result) 
```

# 5) Count vowels and consonants in a string.

```python
x = input("Enter any word or sentence : ")

string = x.lower()
v_count = 0
c_count = 0

for i in string :
    if (i != ' ') :
        if (i=='a' or i=='e' or i=='i' or i=='o' or i=='u') :
            v_count += 1
        else :
            c_count += 1

print("Vowels count :",v_count)
print("Consonants count :",c_count)
```

# 6) Reverse a string without using slicing.

```python
string = input("Enter a String : ")
print()

lst = list(string)
n = len(lst)
lst2 = []

for i in range(n,0,-1):
    lst2.append(lst[i-1])
   
rev_string = ''.join(lst2)

print("After Reversing the string : ",rev_string)
```

# 7) Remove duplicate elements from a list without using set().

```python
lst = ['a', 1, 1, 'd', 'a', 4, 65, 'd', 4, 90, 10, 11, 4]

lst2 = []
for i in lst :
    if i not in lst2:
        lst2.append(i)
    else:
        continue

print("Original list :")
print(lst)
print()
print("Unique list after removing duplicates : ",)
print(lst2)
```

# 8) Merge two dictionaries into one.

```python
dic1 = { 'Name':'Max', 'Age':25, 'Country':'America' }
print("Dictionary 1 : ", dic1)

dic2 = { 'Sex':'Male', 'id':123 }
print("Dictionary 2 : ", dic2)
print()

# option 1
new_dic = dic1|dic2
print("After Merging both dictionaries :")
print(new_dic)
print()

# option 2
new_dic2 = {**dic1,**dic2}
print("After Merging both dictionaries (by another way) :")
print(new_dic2)
```

# 10)  Create a class Student that stores name, marks & displays percentage.

```python
class student:
    def __init__(self,name, marks):
        self.name = name
        self.marks = marks
   
    def score(self):
        per = (self.marks/500)*100
        print("Name : ", self.name)
        print("Marks obtained : ", self.marks)
        print("Percentage : ", per)

obj = student('Max',300)
obj.score()
```

# 11) Sort a list of tuples by the second element.

```python
lst = [(5,3), (2,1), (9,5), (6,2), (1,4)]

lst2 = []
for i in lst:
    lst2.append(i[1])

lst3 = sorted(lst2)
new_lst = []

for i in lst3:
    for j in lst:
        if j[1]==i:
            new_lst.append(j)

print("Original list :")
print(lst)
print("After sorting by second element (ascending order) : ")
print(new_lst)
```

```python
# by using lambda function

lst = [(5,3), (2,1), (9,5), (6,2), (1,4)]

sorted_lst = sorted( lst, key = lambda x: x[1] )

print("Original list :")
print(lst)
print("After sorting by second element (ascending order) : ")
print(new_lst)
```
