```python
# 9) Replace all elements < 5 with 0 in a NumPy array.
```

```python
arr = np.random.randint(1,10,9).reshape(3,3)
print("Array : ")
print(arr)
print()

for i in range(3):
    for j in range (3):
        if arr[i][j] < 5 :
            arr[i][j] = 0
print("Replace all elements less then 5 with 0 :")
print(arr)
```

```python
# 10) Multiply each row of a matrix by a different number.
```

```python
arr = np.ones((4,4))
print("Array : ")
print(arr)
print()

n = 3
for i in range(4):
    for j in range (4):      
        arr[i][j] = arr[i][j] * n
    n = n+1    

print("Multiply each row of a matrix by a different number :")
print(arr)
```

```python
# 10) Convert a column to datetime and extract year, month and day.
```

```python
data = { 'Date': ['2023-01-15', '2022-07-10', '2024-03-25'] }

df = pd.DataFrame(data)
df
```

```python
# Convert to datetime
df['Date'] = pd.to_datetime(df['Date'], format="%Y-%m-%d")          # if there's like 2023/01/15 - "%Y/%m/%d"
df                                                                  # if there's like 2023/january/15 - "%Y/%b/%d" (use b instead of m)           
```

```python
# Extract year and month and day
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
df
```

```python
# 11) Replace similar string data in a column 
```
