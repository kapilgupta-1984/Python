# 10) Cat Plot ( by this we can draw all graphs by 'kind' )

```python
sns.catplot(x = "size", y = "tip", data = df, hue = "sex", 
            kind = "swarm", palette = "bwr")
# kind`:Options are 'strip', 'swarm', 'box', 'boxen', 'violin', 'bar', 'count', and 'point'."
plt.show()
```

# 11) Facet Grid ( to plot multiple plots)

```python
fg = sns.FacetGrid(data=df, col = "sex", hue = "day", palette = "bwr")
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

# 12) Pair Plot

```python
sns.pairplot(data = df, hue = "sex", kind = "reg", diag_kind = "hist")
# "scatter" "kde" "hist" "reg"
plt.show()
```

# Comparing any two plots

```python
plt.subplot(1,2,1)
sns.violinplot(x='day',y='total_bill', data=df, hue='time')

plt.subplot(1,2,2)
sns.boxplot(x="day", y="total_bill", data = df, hue = "time")
plt.show()
```

# 13) Use fill_between() to shade area under a curve.

```python
# Plot the curve
x = df['day']
y = df['time']
plt.plot(x,y, label='tip')

# Shade area under the curve
plt.fill_between(x,y, color='skyblue', alpha=0.4)

plt.title("Shaded Area Under the Curve (tip)")
plt.xlabel("day")
plt.ylabel("time")
plt.legend()
plt.show()
```

# Build a complete dashboard-like figure with subplots:
                                                       # line chart
                                                    # bar chart
                                                        # heatmap
                                                    # scatterplot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set(style="whitegrid")

fig = plt.figure(figsize=(16,10))

# 
# 1 Line Chart  total_bill vs tip
# 
ax1 = plt.subplot(2, 2, 1)
ax1.plot(df['total_bill'], df['tip'], marker='o', linewidth=1)
ax1.set_title("Line Chart: Total Bill vs Tip")
ax1.set_xlabel("Total Bill")
ax1.set_ylabel("Tip")

# 
# 2 Bar Chart  Average tip per day
# 
ax2 = plt.subplot(2, 2, 2)
sns.barplot(x='day', y='tip', data=df, estimator=np.mean)
ax2.set_title("Bar Chart: Average Tip by Day")
ax2.set_xlabel("Day")
ax2.set_ylabel("Avg Tip")

# 
# 3 Heatmap  Correlation Matrix
# 
ax3 = plt.subplot(2, 2, 3)
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
ax3.set_title("Heatmap: Correlation Matrix")

# 
# 4 Scatterplot  total_bill vs tip (hue=sex)
# 
ax4 = plt.subplot(2, 2, 4)
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=df, s=70)
ax4.set_title("Scatter Plot: Total Bill vs Tip")
ax4.set_xlabel("Total Bill")
ax4.set_ylabel("Tip")

plt.tight_layout()
plt.show()
```
